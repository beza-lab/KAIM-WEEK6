```python
import pandas as pd

# Load the data
df = pd.read_csv('D:/week6 data/data.csv')

# Show the dataset dimensions
print(f'Dataset Dimensions: {df.shape}')

# Show the data types of each column
print(f'Data Types:\n{df.dtypes}')

# Display the first few rows of the dataset
print(f'First Few Rows:\n{df.head()}')
```

```python
# Summary statistics for numerical features
print(f'Summary Statistics:\n{df.describe()}')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

numerical_features = ['CountryCode', 'Amount', 'Value', 'PricingStrategy', 'FraudResult']

plt.figure(figsize=(20, 15))
for i, col in enumerate(numerical_features):
    plt.subplot(len(numerical_features)//3 + 1, 3, i + 1)
    sns.histplot(df[col].dropna(), kde=True)
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



# Load the dataset
df = pd.read_csv('D:/week6 data/data.csv')

categorical_features = ['TransactionId', 'BatchId', 'AccountId', 'SubscriptionId', 'CustomerId', 'CurrencyCode', 'ProviderId', 'ProductId', 'ProductCategory', 'ChannelId']

plt.figure(figsize=(20, 15))
for i, col in enumerate(categorical_features):
    plt.subplot(len(categorical_features)//3 + 1, 3, i + 1)
    sns.countplot(x=col, data=df, order=df[col].value_counts().index)
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('D:/week6 data/data.csv')

# Select only numerical columns for correlation analysis
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns
correlation_matrix = df[numerical_columns].corr()

# Plot the correlation matrix
plt.figure(figsize=(15, 10))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
missing_values = df.isnull().sum()
print(f'Missing Values:\n{missing_values}')

sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```

```python
# Define numerical features
numerical_features = ['CountryCode', 'Amount', 'Value', 'PricingStrategy', 'FraudResult']

# Box plots for outlier detection
plt.figure(figsize=(20, 15))
for i, col in enumerate(numerical_features):
    plt.subplot(len(numerical_features)//3 + 1, 3, i + 1)
    sns.boxplot(x=df[col])
    plt.title(f'Outliers in {col}')
plt.tight_layout()
plt.show()
```
